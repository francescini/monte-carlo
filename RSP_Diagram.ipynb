```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Function for the diagram with min, median, and max lines
def create_risk_capability_plot(dangerous_capabilities_min, dangerous_capabilities_median, dangerous_capabilities_max,
                                protective_measures_min, protective_measures_median, protective_measures_max,
                                title=""):
    
    # Create figure and axis
    fig, ax = plt.subplots(figsize=(8, 8))

    # Define the "safe" and "risky" regions
    safe_zone = Polygon([[0, 0], [0, 1], [1, 1], [1, 0]], closed=True, facecolor='lightblue', alpha=0.4)
    risky_zone = Polygon([[0.01, 0], [1, 0], [1, 1], [1, 1]], closed=True, facecolor='lightcoral', alpha=0.4)

    ax.add_patch(safe_zone)
    ax.add_patch(risky_zone)

    # Plot lines for min, median, and max scenarios
    ax.plot(dangerous_capabilities_min, protective_measures_min, marker='o', linestyle='--', color='red', label='Worst-case (Min)')
    ax.plot(dangerous_capabilities_median, protective_measures_median, marker='o', linestyle='-', color='gray', label='Median-case')
    ax.plot(dangerous_capabilities_max, protective_measures_max, marker='o', linestyle='--', color='green', label='Best-case (Max)')

    # Add labels and annotations
    ax.set_xlabel('Dangerous Capabilities', fontsize=12)
    ax.set_ylabel('Protective Measures', fontsize=12)
    ax.set_title(title, fontsize=14)

    # Define limits
    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])

    # Add text to different regions
    ax.text(0.25, 0.75, 'Safe region:\nSufficient protection', fontsize=10, verticalalignment='center', horizontalalignment='center', color='blue')
    ax.text(0.75, 0.25, 'Risky region:\nHigh capabilities, insufficient protection', fontsize=10, verticalalignment='center', horizontalalignment='center', color='red')

    # Show grid and legend
    ax.grid(True)
    ax.legend()

    # Display plot
    plt.show()

# Input data
dangerous_capabilities_min = [0.01, 0.2, 0.4, 0.55, 0.7, 0.8, 0.95]
dangerous_capabilities_median = [0.01, 0.2, 0.4, 0.55, 0.7, 0.8, 0.95]
dangerous_capabilities_max = [0.01, 0.2, 0.4, 0.55, 0.7, 0.8, 0.95]

protective_measures_min = [0.02, 0.2, 0.4, 0.5, 0.65, 0.75, 0.8]
protective_measures_median = [0.05, 0.25, 0.45, 0.55, 0.7, 0.8, 0.9]
protective_measures_max = [0.1, 0.3, 0.5, 0.6, 0.75, 0.85, 0.95]

# Call function to plot the diagram with three lines
create_risk_capability_plot(dangerous_capabilities_min, dangerous_capabilities_median, dangerous_capabilities_max,
                            protective_measures_min, protective_measures_median, protective_measures_max)
```
